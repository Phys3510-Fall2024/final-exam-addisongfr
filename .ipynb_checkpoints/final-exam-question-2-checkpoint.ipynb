### The potential energy (notice the minus sign on the fourth-order term)

$$ \Delta U = \frac{1}{2} K Q^2 - \frac{1}{4} D Q^4 + \frac{1}{6} \Phi Q^6 $$

### Plot of the potential energy

```python
import numpy as np
import matplotlib.pyplot as plt

# define potential function
def potential_energy(Q, K=1.0, D=2.4, Phi=1.0):
    return (1/2) * K * Q**2 - (1/4) * D * Q**4 + (1/6) * Phi * Q**6

# define mesh for plotting
Q_mesh = np.linspace(-2,2,100)
U_mesh = potential_energy(Q_mesh)

# construct figure
plt.figure(figsize=(5, 4))
plt.plot(Q_mesh, U_mesh, color="gray")
plt.xlabel("$Q$")
plt.ylabel("$\Delta U$")
plt.xlim(-2,2)
plt.ylim(-0.1,0.25)
plt.show()
```

### Streamlining Everything

```python
import numpy as np
from scipy.integrate import solve_ivp
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

K = 1
D = 2.4
phi = 1
M = 1

def U(Q, K, D, phi):
    U = (1/2) * K * Q**2 - (1/4) * D * Q**4 + (1/6) * phi * Q**6
    return U
    
def F(Q, K, D, phi):
    F = - (K * Q - D * Q**3 + phi * Q**5)
    return F

def equations_of_motion(t, y):
    Q, v = y                      
    dQ_dt = v                        
    dv_dt = F(Q, K, D, phi)          
    return [dQ_dt, dv_dt]

def find_period(Q_0, t_max = 50, points = 1000):
    t_span = (0, 50)
    t_eval = np.linspace(*t_span, points)
    sol_period = solve_ivp(equations_of_motion, t_span, [Q_0, 0], t_eval = t_eval, method = 'RK45')
    Q = sol_period.y[0]
    t = sol_period.t

    maxima = np.where((Q[1:-1] > Q[:-2]) & (Q[1:-1] > Q[2:]))[0] + 1  
    if len(maxima) > 1:                                                
        period = t[maxima[1]] - t[maxima[0]]
        return period
    else:
        return np.nan

Q_0_vals = np.linspace(-1.7, 1.7, 50)  # Initial position given
freq_vals = []
for Q_0 in Q_0_vals:
    period = find_period(Q_0)
    if period:
        freq_vals.append(2 * np.pi / period)
    else:
        freq_vals.append(np.nan)
freq_vals = np.array(freq_vals)

plt.figure(figsize=(8, 6))
plt.plot(Q_0_vals, freq_vals)
plt.title('Angular Frequency vs. Initial Displacement for Anharmonic Oscillator')
plt.xlabel('Initial Displacement $Q_0$')
plt.ylabel('Angular Frequency')
plt.grid()
plt.show()
```

### Describing the Plot

The plot above shows the behavior of the angular frequency of the anharmonic oscillator, $\omega$, as a function of its initial displacement, $Q_0$. In the center of the plot, where the displacement of $Q_0$ is small, the system behaves as a simple harmonic oscillator, with vallies of frequency (very few if any oscillations) around displacements of $Q_0$ = -0.75 and $Q_0$ = 0.75. This corresponds to wells in potential energy. However, when $Q_0$ is displaced further than these values, the frequency begins to increase rapidly, peaking at around -1.4 and 1.4. This is where we see the anharmonic terms of the oscillator's potential energy begin to influence the system much more heavily, and the system no longer behaves like a simple harmonic oscillator. Compared to the plot of potential energy, where there are local maxima around -1.7 and 1.7, this is where there are local minima in the plot of initial displacement versus angular frequency.
